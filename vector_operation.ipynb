```python
vectors_1 = np.array([0,0,2,3])

vectors_2 = np.array([0,0,3,-2])

sum =vectors_1+ vectors_2
print(sum)
```

subtraction of vectors

```python
vectors_1= np.array([0,0,2,3])

vectors_2 = np.array([0,0,3,-2])

difference = vectors_1 - vectors_2
print(difference)
```

multiplying of vector

```python
vector_1 = np.array([0,0,2,3])

vector_2 = np.array([0,0,2,2])

multiply = vector_1 * vector_2

print(multiply)
```

```python
# dot product of a vector
x= np.array([2,3])

y = np.array([4,4])

x_dot_y = np.dot(x,y)
print(x_dot_y)
```

```python
z =np.array([40,20,35])

d =np.array([53,24,68])

z_dot_d=np.dot(z,d)
print(z_dot_d)
```

```python
#cross product of a vector
x =np.array([2,3,4])

y = np.array([4,5,6])

x_cross_y = np.cross(x,y)
print(x_cross_y)
```

```python
# projection of vector
a = np.array([2,5])

v = np.array([8,-6])

#magnitude of 'v' vector
magnitude_of_v = np.sqrt(np.sum(v**2))

proj_of_a_on_v = (np.dot(a,v)/magnitude_of_v**2 )*v
print('proj_of_a_on_v:',proj_of_a_on_v)
```
