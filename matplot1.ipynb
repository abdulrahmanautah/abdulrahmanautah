```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
plt.plot(x,y)
plt.show()

print(x)

```

```python
#adding tittle to x axis and y axis
x=np.linspace(0,10,100)
y=np.sin(x)
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sinvale')
plt.title('sine value')
plt.show()


```

```python
#parabola

x=np.linspace(-10,10,50)
y=x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'r+')
plt.show()
```

```python
#multiple lines plot
x=np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'g.')
plt.plot(x,np.cos(x),'b+')
plt.show()


```

```python
#Bar chart
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['english','french','spanish','hausa']
people =[30,50,60,20]
ax.bar(languages,people)
plt.xlabel('languages')
plt.ylabel('people')
plt.show()

```

```python
#piechart
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.pie(people,labels=languages,autopct='%1.1f%%')
plt.show()
```

```python
#scatter plot
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
fig2=plt.figure()
ax=fig2.add_subplot(1,1,1)
ax.scatter(x,y,color = 'g')
ax.scatter(x,z,color ='r')
plt.show()
```

```python
#3D scatter plot
fig=plt.figure()
x =plt.axes(projection= '3d')
z=20*np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```
