```python
#Data visualization library

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

#note seaborn has some built in datasets
# 1.tips dataset
# total vs tip datasets

tips= sns.load_dataset('tips')
tips.head()

```

# visualize the data



```python
#sns.relplot(tips)
sns.set_theme()
sns.relplot(data=tips, x='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.relplot(x='sepal_length',hue='species', y='petal_width',data=iris)
```

```python
sns.relplot(x='petal_length',hue='species', y='sepal_width',data=iris)
```

```python
#laoding the titanic dataset

titanic=sns.load_dataset('titanic')
```

```python
#count plot
sns.countplot(x='class',data=titanic)
```

```python
#bar chart
sns.barplot(x='sex',y='survived',hue='class',data=titanic)
```

```python
from sklearn.datasets import load_diabetes
diabetes_dataset = load_diabetes()
diabetes = pd.DataFrame(diabetes_dataset.data,columns=diabetes_dataset.feature_names)
diabetes['price'] = diabetes_dataset.target
diabetes.head()
```

```python
sns.displot(diabetes['price'])
```

constructing heatmap

```python
correlation = diabetes.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Blues')
plt.show()
```
